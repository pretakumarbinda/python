```python
days = [1,2,3,4,5,6]
max_t = [50,51,52,53,54,48]
min_t = [42,41,43,30,38,33]
avg_t = [42,41,43,40,35,33]

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.xlabel('Days')
plt.legend()  #indicates max,min,avg wrt days
```

```python
#legend er location change korte chaile:

days = [1,2,3,4,5,6]
max_t = [50,51,52,53,54,48]
min_t = [42,41,43,30,38,33]
avg_t = [42,41,43,40,35,33]

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.xlabel('Days')
plt.legend(loc = "upper right")



```

```python
product = [1,2,3,4]
price = [15,25,40,30]
plt.bar(product, price, label='price')
```

```python
product = [1,2,3,4]
price = [15,25,40,30]
plt.bar(product, price, label='price', width=0.5, color = 'red')
```

```python
product = [1,2,3,4]
price = [15,25,40,30]
profit=[5,7,10,5]
plt.bar(product, price, label='price', color = 'red')
plt.bar(product, profit, label='profit')
```

```python
import numpy as np
```

```python
product = np.array([1,2,3,4])
price = [15,25,40,30]
profit=[5,7,10,5]
plt.bar(product-.2, price, label='price', color = 'red')
plt.bar(product, profit, label='profit')
```

```python
product = np.array([1,2,3,4])
price = [15,25,40,30]
profit=[5,7,10,5]
plt.bar(product-.2, price, label='price', color = 'red', width=.4)
plt.bar(product+.3, profit, label='profit',width = .4)
plt.legend()
```

```python
val = [300,500,1600,400,50,360]
lab= ['price', 'sell','Total','Shirt', 'sari', 'juta']
```

```python
plt.pie(val)
```

```python
plt.pie(val, labels = lab, radius=2)
```

```python
#to show percent:
plt.pie(val, labels = lab, autopct='%.2f%%')
```

```python

plt.pie(val, labels = lab, autopct='%.2f%%', shadow=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

plt.pie(val, labels = lab, autopct='%.2f%%', shadow=True, startangle=45)  #first value 'price' will end at 45 degree
```
